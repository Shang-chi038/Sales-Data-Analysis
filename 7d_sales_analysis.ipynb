```python
# Confirming I have pandas installed

import pandas as pd
print(pd.__version__)
```

```python
# Load Sales Data

sales_df = pd.read_csv('data/sample_superstore.csv')
print(sales_df.head())
```

```python
# Group Data by Product Category

grouped_sales = sales_df.groupby('Category')
```

```python
# Calculate Total Sales for Each Category

total_sales_by_category = grouped_sales['Sales'].sum()
```

```python
# Display Results

print(total_sales_by_category)
```
