```python
# Import Libraries and Load Data

import pandas as pd
import matplotlib.pyplot as plt

# Load the sales data
sales_data = pd.read_csv('sales_data_for_9.csv')

# Print the head of the DataFrame
print(sales_data.head())
```

```python
# Aggregate Sales Data by Month

import datetime as dt

# Convert 'Date' to datetime
sales_data['Date'] = pd.to_datetime(sales_data['Date'])

# Extract month
sales_data['Month'] = sales_data['Date'].dt.month

# Aggregate sales by month
montly_sales = sales_data.groupby('Month')['Sales'].sum()

print(montly_sales)
```

```python
# Create a Line Chart of Monthly Sales

plt.plot(montly_sales.index, montly_sales.values, marker='o', linestyle='-')
plt.show()
```

```python
# Add Title and Axis Labels

plt.plot(montly_sales.index, montly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (NGN)')
plt.show()
```

```python
# Aggregate Sales Data by Product Category

category_sales = sales_data.groupby('Product')['Sales'].sum()
print(category_sales)
```

```python
# Create a Bar Chart of Sales by Category

plt.figure(figsize=(10, 6))
plt.bar(category_sales.index, category_sales.values)
plt.title('Sales by Product Category')
plt.ylabel('Sales (NGN)')
plt.xlabel('Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
