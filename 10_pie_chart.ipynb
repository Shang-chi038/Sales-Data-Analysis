```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Sales Data from CSV

sales_data = pd.read_csv('sales_data_for_9.csv')
print(sales_data.head())
```

```python
# Aggregate Sales by Products

sales_by_product = sales_data.groupby('Product')['Sales'].sum()#.reset_index()
# if using pie, don't reset index
print(sales_by_product.info())
print(type(sales_by_product)) # note 'sales_by_product' is not a dataframe, rather it is a series, because it was derived from a Dataframe
```

```python
# Create the Pie Chart

plt.figure(figsize=(8, 8))
plt.pie(sales_by_product, labels=sales_by_product.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by Product Category')

# Save the Pie Chart to a File
plt.savefig('reports/sales_by_category_.png')
# do not show a pie chart before saving, e get why, dem no dey tell egugu twice
```
