```python
import pandas as pd
print(pd.__version__)
```

```python
# Recreating sales_data.csv
from random import randint
with open('sales_data.csv', 'w', newline='') as file:
    file.write("product_name,sales_figure\n") # header

    for i in range(65, 91): #  I'll use chr and create product_name and randint for sales_figures
        product = 'Product ' + chr(i)
        sales = randint(100, 900)
        file.write(f"{product},{sales}\n")
```

```python
# 1 - Load Sales Data

sales_data = pd.read_csv('sales_data.csv')
print(sales_data.head()) # this returns first five elements by default
```

```python
# 2 - Selecting a Single Row Using `.loc` with a Label

sales_data = sales_data.set_index('product_name')
selected_row = sales_data.loc['Product O']
print(selected_row)
```

```python
# 2b - Selecting a Single Row Using `.iloc` with a Label

selected_row = sales_data.iloc[10]
print(selected_row)
```

```python
# 3 - Selecting Multiple Rows Using `.loc` with a List of Labels

selected_rows = sales_data.loc[['Product M','Product P']] # take note of the quotation marks around the two indeces
print(selected_rows)
```

```python
print(sales_data.index)
```

```python
# after removing index 

sales_data = sales_data.reset_index()
selected_rows = sales_data.loc[['Product M','Product P']] # take note of the quotation marks around the two indeces
print(selected_rows)
```

```python
print(sales_data.index) # take note of the difference
```

```python
# 5 - Selecting a Single Row Using `.iloc` with an Integer Position

sales_data = sales_data.reset_index()
selected_row = sales_data.iloc[2]
print(selected_row)
```

```python
# 6 - Selecting Multiple Rows Using `.iloc` with a List of Integer Positions

selected_rows = sales_data.iloc[[1, 3, 5]]
print(selected_rows)
```

```python
# 7 - Selecting a Range of Rows Using `.iloc` with Slicing

selected_rows = sales_data.iloc[2:5]
print(selected_rows)
```

```python
# 7b - Selecting a Range of Rows Using `.loc` with slicing

sales_data = sales_data.set_index('product_name')
selected_rows = sales_data['Product M' : 'Product P']
print(selected_rows)
```

```python
sales_data = sales_data.reset_index()
print(sales_data.index)
```
