```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Sales Data

sales_data = pd.read_csv('data/sample_superstore.csv')
print(sales_data.head())
```

```python
# Aggregate Region by Sales

sales_by_region = sales_data.groupby('Region')['Sales'].sum().reset_index()
print(sales_by_region)
```

```python
# Create a bar chart using subplots

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(sales_by_region['Region'], sales_by_region['Sales'], color='skyblue')
ax.set_xlabel('Region')
ax.set_ylabel('Total Sales')
ax.set_title('Total Sales by Region')

# Rotate X-Axis Labels
plt.xticks(rotation=30, ha='right')
```

```python
# Create a bar chart using bar()
plt.figure(figsize=(10, 6))
plt.bar(sales_by_region['Region'], sales_by_region['Sales'], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Total Sales by Region')

# Rotate X-Axis Labels
plt.xticks(rotation=30, ha='left')
```
