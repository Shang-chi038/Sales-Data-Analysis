```python
# Confirming I have pandas and numpy installed

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
# Creating the file I will use for this task

header_rows = ['TransactionID','Date','Product','SalesAmount']
data_rows = [[1,'2024-01-01','ProductA',100],
            [2,'2024-01-01','ProductB',150],
            [3,'2024-01-02','ProductA',200],
            [4,'2024-01-02','ProductC',50],
            [5,'2024-01-03','ProductB',120],
            [6,'2024-01-03','ProductC',80],
            [7,'2024-01-04','ProductA',180],
            [8,'2024-01-04','ProductB',130],
            [9,'2024-01-05','ProductC',70],
            [10,'2024-01-05','ProductA',220]]

df = pd.DataFrame(data_rows, columns=header_rows)
df.to_csv('sales_data_6.csv', index=False)
```

```python
# Loading the data

sales_data = pd.read_csv('sales_data_6.csv')
print(sales_data)
```

```python
# Calculate the total Sales

sales_amounts = sales_data['SalesAmount']
total_sales = np.sum(sales_amounts)

print(f'Total Sales: {total_sales}')
```

```python
# Calculate Average Sales

sales_amounts = sales_data['SalesAmount']
average_sales = np.mean(sales_amounts)

print(f'Average Sales: {average_sales}')
```

```python
# Clculate the Standard Deviation of Sales

sales_amounts = sales_data['SalesAmount']
std_dev_sales = np.std(sales_amounts)

print(f'Standard Deviation of Sales: {std_dev_sales}')
```
