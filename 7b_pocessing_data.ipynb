```python
# 1 - Confirming I have pandas installed

import pandas as pd
print(pd.__version__)
```

```python
# 2 -Defining the Filtering Function

def filter_by_region(df, region_name):
    """Filters the DataFrame to include only sales records from the specified region."""
    filtered_df = df[df['Region'] == region_name]
    return filtered_df
```

```python
# 3 - Defining the Load Sales Data Function

def load_sales_data(file_path):
    """Loads sales data from a CSV file into a Pandas DataFrame."""
    df = pd.read_csv(file_path)
    return df
```

```python
# 4 - Apply the Filtering Function

file_path = 'data/sample_superstore.csv'
sales_data = load_sales_data(file_path)
region_name = 'Central'
filtered_data = filter_by_region(sales_data, region_name)
```

```python
# 5 - Print Filtered Data

print(filtered_data.head())
```

```python
# 6 - Save the Filtered Data 

filtered_data.to_csv('data/central_sales_data.csv', index=False)
```
